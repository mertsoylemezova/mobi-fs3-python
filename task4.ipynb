**<a id="#">Task 4.1.1.</a>** Load the image `data/astronaut.png` and show it:

**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python

 h_mask = [
    [-1,0,1],
    [-1,0,1],
    [-1,0,1]
 ]

 v_mask = [
    [-1,-1,-1],
    [0,0,0],
    [1,1,1]
 ]
def prewitt_h(img):
    img2=img.copy()
    result = numpy.zeros(img2.shape)
    
    for p in  numpy.ndindex(img2.shape):
        if 1 <= p[0] < img2.shape[0] - 1 and 1 <= p[1] < img2.shape[1] - 1:
            numpy.sum(img2[p[0]-1:p[0]+2,p[1]-1:p[1]+2]*h_mask)/6
            result[p]= numpy.sum(img2[p[0]-1:p[0]+2,p[1]-1:p[1]+2]*h_mask)/6
    return result


def prewitt_v(img):
    img2=img.copy()
    result = numpy.zeros(img2.shape)
    
    for p in numpy.ndindex(img2.shape):
        if 1 <= p[0] < img2.shape[0] - 1 and 1 <= p[1] < img2.shape[1] - 1:
            numpy.sum(img2[p[0]-1:p[0]+2,p[1]-1:p[1]+2]*v_mask)/6
            result[p]= numpy.sum(img2[p[0]-1:p[0]+2,p[1]-1:p[1]+2]*v_mask)/6
    return result
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
plt.figure(figsize=(15,8))
plt.imshow(prewitt_h(img))

plt.figure(figsize=(15,8))
plt.imshow(prewitt_v(img))
```
