## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/astronaut.png` and show it:

```python
img = plt.imread('data/astronaut.png')
plt.imshow(img, 'gray')
plt.colorbar()
```

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img, size):
    img2=img.copy()
    result = numpy.zeros(img2.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img2.shape):
        neighborhood_sum = 0
        if n <= p[0] < img2.shape[0] - n and n <= p[1] < img2.shape[1] - n:
            for q in numpy.ndindex(size, size):
                neighborhood_sum += img2[(p[0] - n) + q[0], (p[1] - n) +q[1]]
            result[p] = neighborhood_sum / size**2
            
    return result
img_mean = meanfilter(img,5)
plt.imshow(img_mean)
```
